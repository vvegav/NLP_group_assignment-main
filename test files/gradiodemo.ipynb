```python
%%blocks
css = """
h1 {
    text-align: center;
    display:block;
}
"""
with gr.Blocks(css=css) as demo:
  # First Section - Twitter News and Summary
  with gr.Group():
    gr.Markdown("# Trendy News Dashboard")

  with gr.Row():
    gr.Button('Tweet1')
    gr.Button('Tweet2')
    gr.Button('Tweet3')
  with gr.Row():
    gr.Button('Tweet4')
    gr.Button('Tweet5')
    gr.Button('Tweet6')
  with gr.Row():
    gr.Button('Tweet7')
    gr.Button('Tweet8')
    gr.Button('Tweet9')

  generate_summary_btn = gr.Button("Generate Summary")
  #generate_summary_btn.click(fn= ...) ADD FUNCTION TO GENERATE
  summary_radio = gr.Radio(summary_options, label="Select your Summary")
  select_summary_btn = gr.Button('Select your Summary')
  select_summary_btn.click(fn= summary_selection, inputs= summary_radio, outputs= gr.Textbox(label= 'Selected Summary'))


  # Second Section - Upload Video and Transcript
  with gr.Group():
    gr.Markdown("# Upload your Video")
    gr.Markdown("### Upload your video so we can extract your style and help you generate new content!")

  with gr.Row():
    video_upload = gr.Video(label="Upload a Video")
    video_transcript = gr.Textbox(label ='Video Transcript')

  video_upload_btn = gr.Button ("Upload your Video")
  video_upload_btn.click(fn=transcribe_video, inputs=video_upload, outputs= video_transcript)

  # Third Section - Generation and Selection of Script
  with gr.Group():
    gr.Markdown("# Scirpt Generation")
    gr.Markdown("### Now we will generate some options of a script for your new video. Choose your favorite!")

  generate_script = gr.Button('Generate Scripts')
  #generate_script.click(fn = ...) #ADD FUNCTION TO GENERATE SCRIPTS

  with gr.Row():
    script_radio = gr.Radio(script_options, label="Select your Script")

  select_script_btn = gr.Button('Select your Script')
  select_script_btn.click(fn= script_selection, inputs= script_radio, outputs= gr.Textbox(label= 'Final Script'))

# demo.launch(server_port=7889)
```
